```python
import subprocess

# List of arguments you want to loop through
methods_to_test = ["FSNet", "DC3"]

for method in methods_to_test:
    print(f"--- Running main.py with method: {method} ---")
    
    # This runs the command just like in the terminal
    subprocess.run([
            "python", "main.py",
            "--method", method,
            "--prob_type", "convex",
            "--prob_name", "qp",
            "--network", "MLP",
            "--train_size", "2000",
            "--activation", "ReLU"
        ])
    
print("All runs completed.")
```
